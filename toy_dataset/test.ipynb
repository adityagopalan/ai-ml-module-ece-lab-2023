# Load data

```python
X1= np.genfromtxt("./class_A/data.csv", delimiter=',')
X2= np.genfromtxt("./class_B/data.csv", delimiter=',')
```

# PCA

```python
X=np.vstack((X1,X2))
pca_1 = PCA(n_components=2)
X1_hat=pca_1.fit_transform(X)
plt.scatter(X1_hat.T[0],X1_hat.T[1],c='blue')
plt.show()
```
